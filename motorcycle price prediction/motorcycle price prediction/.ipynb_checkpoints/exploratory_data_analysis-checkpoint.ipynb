## 1. Phân tích dữ liệu định lượng

Dữ liệu định lượng là loại dữ liệu có thể được đo lường và biểu diễn thông qua số lượng, bao gồm bất kỳ thông tin nào có thể được đếm hoặc đo lường. Các trường dữ liệu định lượng này có thể được đánh giá thông qua các tham số thống kê như giá trị trung bình, giá trị trung vị, tứ phân vị,...

Các trường dữ liệu định lượng trong bộ dữ liệu gồm:
<ul>
  <li>Tuoi_xe</li>
  <li>So_km_da_di</li>
  <li>Gia_xe</li>
</ul>

```python
# Xem các tham số thống kê
df[["Tuoi_xe", "So_km_da_di", "Gia_xe"]].describe()
```

Cột So_km_da_di có trung bình 28,946 km, nhưng trung vị chỉ 12,000 km, cho thấy phần lớn xe máy có quãng đường di chuyển thấp (từ 3,000 đến 25,000 km), phù hợp với xe phổ thông sử dụng hàng ngày. Tuy nhiên, một số ít xe có quãng đường rất lớn (lên đến 999,999 km)

Cột Gia_xe có trung bình 32,38 triệu VND, nhưng trung vị chỉ 18,5 triệu VND, cho thấy phần lớn xe có giá thấp (10-32.8 triệu VND), nhưng một số xe có giá rất cao (lên đến 36.5 tỷ VND)

Từ đó nhận thấy dữ liệu lệch phải mạnh ở So_km_da_di và Gia_xe, do các outliers (999,999 km, 36,5 tỷ VND)

Độ lệch chuẩn lớn cũng cho thấy dữ liệu có nhiều giá trị bất thường (Gia_xe: 317 triệu VND, So_km_da_di: 80,673 km)

Các giá trị bất thường này cần được kiểm tra bằng biểu đồ để đánh giá tác động

Độ lệch chuẩn lớn (So_km_da_di: 80,673 km; Gia_xe: 317 triệu VND) xác nhận dữ liệu phân tán mạnh, không ổn định và cần xử lý giá trị ngoại lai để cải thiện

## 2. Phân tích dữ liệu định tính

Dữ liệu định tính là loại dữ liệu không thể được đo lường bằng số mà thường được mô tả bằng ngôn từ. Nó chủ yếu dùng để mô tả đặc điểm, thuộc tính, hoặc hiện tượng

```python
df.columns
```

```python
# Kiểm tra số lượng các hãng xe có trong bộ dữ liệu
df["Hang_xe"].value_counts()
```

Các hãng xe có số lượng thấp thường sẽ rất khó đánh giá tính chính xác khi xây dựng mô hình, nên ta sẽ gom các hãng xe có số lượng ít, cụ thể ta có thể đặt ra ngưỡng số lượng dưới 50 thành giá trị "Hãng khác"

```python
# Đổi các hãng xe có số lượng xe dưới 50 thành hãng khác
df['Hang_xe'] = df['Hang_xe'].apply(\
                lambda x : 'Hãng khác' if len(df[df['Hang_xe']==x]) < 50 else x)

df["Hang_xe"].value_counts()
```

```python
def plot_bar(column):
  sns.countplot(data=df, x=column, order=df[column].value_counts().index)
  plt.xticks(rotation=45)

  plt.show()

plot_bar('Hang_xe')
```

Biểu đồ cột cho thấy hãng xe Honda dẫn đầu thị trường xe máy Việt Nam, tiếp theo là Yamaha, Piaggio. Hãng khác có số lượng đáng kể nhưng thấp hơn các hãng chính

Lưu ý: Nếu trường thuộc tính Hãng xe có giá trị Hãng khác, còn Dòng xe lại có giá trị cụ thể, thì rất có thể các dòng xe này đã được có giá trị hãng xe tương ứng ở các dòng dữ liệu này, cần kiểm tra và xem xét

```python
hang_va_dong =  df[['Hang_xe', 'Dong_xe']]
hang_va_dong.drop_duplicates(inplace=True)
hang_va_dong.head(10)
```

```python
pd.set_option('display.max_rows', None)

hang_va_dong[hang_va_dong['Dong_xe'].duplicated(keep=False)]
```

Nhận thấy, một số dòng xe (Cub, Scoopy, Giorno, GT) xuất hiện ở cả Hãng khác và các hãng cụ thể (Honda, Piaggio)

Theo thông tin từ Chợ tốt xe:

Cub, Scoopy, Giorno thuộc Honda

GT thuộc Piaggio (dòng Vespa GT)

```python
df["Hang_xe"] = np.where(df["Dong_xe"] == "Cub", 'Honda', df['Hang_xe'])
df["Hang_xe"] = np.where(df["Dong_xe"] == "Scoopy", 'Honda', df['Hang_xe'])
df["Hang_xe"] = np.where(df["Dong_xe"] == "Giorno", 'Honda', df['Hang_xe'])
df["Hang_xe"] = np.where(df["Dong_xe"] == "GT", 'Piaggio', df['Hang_xe'])

df.head(10)
```

```python
# Giá trung bình theo Dong_xe
gia_TB = df.groupby(['Hang_xe', 'Dong_xe'], as_index=False)['Gia_xe'].mean()
cac_hang = list(set(df['Hang_xe']))

for hang in cac_hang:
    print(f"Hãng xe: {hang}")
    temp_df = gia_TB[['Dong_xe', 'Gia_xe']][gia_TB['Hang_xe'] == hang]
    sns.barplot(data=temp_df, x="Dong_xe", y="Gia_xe")
    plt.xticks(rotation=90)

    plt.show()
```

Biểu đồ cột giá trung bình theo Dong_xe cho thấy sự khác biệt rõ rệt về giá giữa các hãng, có sự chênh lệch khá nhiều về giá trung bình. Đồng thời, Dong_xe luôn thuộc về một Hang_xe, do đó khi xây dựng mô hình dự báo về giá xe, ta có thể loại đi thuộc tính về Hang_xe

```python
# Tinh_trang_xe, Dung_tich_xe, Loai_xe
con_lai = ['Tinh_trang_xe', 'Dung_tich_xe', 'Loai_xe']

def show_value_counts(col):
    plot_bar(col)
    print('Cụ thể')
    print(df[col].value_counts())

for col in con_lai:
    print(f"Biểu đồ về số lượng của cột {col}:")
    show_value_counts(col)
```

Các biểu đồ cột về số lượng cho thấy:

Phần lớn các xe là xe đã sử dụng, tuy các xe mới là không nhiều, nhưng nó vẫn có thể đóng góp vào mô hình do cũng có tới 185 dữ liệu về các xe mới

Các xe được bán phần lớn là xe có phân khối từ 100-175cc

Xe tay ga được bán nhiều hơn so với xe tay côn hay xe số


```python
# Giá xe trung bình theo Dung_tich_xe và Loai_xe:
def show_price(col):

  gia_TB = df.groupby([col], as_index=False)['Gia_xe'].mean()
  print(f"Giá xe trung bình theo thuộc tính: {col}")
  sns.barplot(data=gia_TB, x=col, y="Gia_xe")
  plt.xticks(rotation=90)

  plt.show()

show_price('Dung_tich_xe')
show_price('Loai_xe')
```

```python
# Giá xe trung bình theo Tinh_trang_xe:
df_moi = df[['Tinh_trang_xe', 'Dong_xe', 'Gia_xe']][df['Tinh_trang_xe'] == 'Mới']
df_cu = df[['Tinh_trang_xe', 'Dong_xe', 'Gia_xe']].\
            loc[(df['Tinh_trang_xe'] == 'Đã sử dụng') & \
                df['Dong_xe'].isin(df_moi['Dong_xe'])]


frames = [df_moi, df_cu]

df_tinh_trg = pd.concat(frames)

gia_TB = df_tinh_trg.groupby('Tinh_trang_xe', as_index=False)['Gia_xe'].mean()
print(f"Giá xe trung bình theo thuộc tính: {'Tinh_trang_xe'}")
sns.barplot(data=gia_TB, x='Tinh_trang_xe', y="Gia_xe")
plt.xticks(rotation=90)

plt.show()
```

Các biểu đồ cột về giá trung bình cho thấy:

Xe tay ga và xe tay côn có giá trung bình cao hơn so với xe số
 
Xe phân khối càng cao thì giá càng cao (Tỷ lệ thuận)

Xe mới có giá cao hơn so với xe cũ cùng loại
